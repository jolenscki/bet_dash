```python
import pandas as pd
import re

XLSX = "bet experiment.xlsx"
```

```python
df = pd.read_excel(XLSX, usecols='A:F')
```

```python
df = df[df['Game'] == 'CS:GO'].reset_index(drop=True)
df = df.replace(r"^ +| +$", r"", regex=True)
df
```

```python
rgx = r"(.+) vs (.+)"
df[['team1', 'team2']] = df.pop('Context').str.split(rgx, expand=True)[[ 1, 2]]
```

```python
rgx = r"win \((.+)\)"
df[['winner']] = df.pop('Booking').str.split(rgx, expand=True)[[1]]
```

```python
df['winner'] = df.apply(lambda row: {row.team1: 1, row.team2: 2}[row.winner], axis=1)
```

```python
df = df.drop("Game", axis=1)
```

```python
df = df.rename(columns={'Bet': 'bet_amount', 'Odd': 'odd'})
```

```python
df['result'] = df.apply(lambda row: 1 if row.Result == 0 else 2, axis=1)
```

```python
df = df.drop('Result', axis=1)
```

```python
df['match_id'] = 1
```

```python
df = df[['team1', 'team2', 'winner', 'odd', 'bet_amount', 'match_id', 'result']]
```

```python
import sqlite3
from sqlite3 import Error

DB_FILE = "bet_history.db"
TABLE = "matches"

```

```python
def create_connection(db_file):
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
    
    return conn

def update_table(_conn, table, df):
    r = df.to_sql(
        table,
        _conn,
        index=False, 
        if_exists='append'
        )

conn = create_connection(DB_FILE)
update_table(conn, TABLE, df)
```

```python
_conn = create_connection(DB_FILE)

query = f'''SELECT bet_id,
    team1,
    team2,
    winner,
    odd,
    bet_amount,
    match_id,
    result
    FROM {TABLE}
'''

df = pd.read_sql_query(query, _conn)
```
